 *	(EXISTING) AT content.
 *	(EXISTING) GC ratio.

```python
## select the best data from the set
errors = abs(np.log2(df_full['transl_r1'])-np.log2(df_full['transl_r2']))
threshold = 0.5
sel_gene = errors < threshold
df_full_sel = df_full[sel_gene]
df_full_sel.to_csv('data_set')


cor_replicates = cor(log2(df_full['transl_r1']),np.log2(df_full['transl_r2']))[0]
cor_replicates_clean = cor(log2(df_full['transl_r1'])[sel_gene],np.log2(df_full['transl_r2'])[sel_gene])[0]


# Plot of all the features
plt.figure(figsize=(18,15))
plt.subplot(221)
plt.scatter(log2(df_full['transl_r1']),np.log2(df_full['transl_r2']),c=errors,s=50,alpha=25)
plt.xlabel('Replicate 1 (log2)',fontsize=20)
plt.ylabel('Replicate 2 (log2) ',fontsize=20)
plt.title('Replicates matching',fontsize=25)
plt.text(6,0,'R = %2f' % cor_replicates,fontsize=20)
plt.colorbar()

plt.subplot(222)
plt.plot(sorted(errors),'b-')
plt.xlabel('Gene',fontsize=20)
plt.ylabel('Abs. difference between replicates (2-fold)',fontsize=20)
plt.title('Gene errors',fontsize=25)
plt.axhline(y=threshold, xmin=0.0, xmax=250, linewidth=1, color = 'k')

plt.subplot(223)
plt.scatter(np.log2(df_full['transl_rate_av']),errors.astype('float64'),c=errors,s=50,alpha=25)
plt.xlabel('Av. Translation rate (log2)',fontsize=20)
plt.ylabel('Abs. difference between replicates (2-fold) ',fontsize=20)
plt.title('Replicate Errors',fontsize=25)
plt.ylim(0, 10 )
plt.colorbar()

plt.subplot(224)
plt.scatter(log2(df_full['transl_r1'])[sel_gene],np.log2(df_full['transl_r2'])[sel_gene],s=50)
plt.xlabel('Replicate 1 (log2)',fontsize=20)
plt.ylabel('Replicate 2 (log2) ',fontsize=20)
plt.title('Replicates matching',fontsize=25)
plt.text(6,0,'R = %2f' % cor_replicates_clean,fontsize=20)
```

## 4. Gaussian process model for the translational efficiency of mammalian cells

<img src="model.jpeg">

```python
# Select training and test data
np.random.seed(123)
Ntrain = 700
Ntest = N - Ntrain
myRange = range(N)
np.random.shuffle(myRange)

# Index of the training and test genes
sel_train = myRange[0:(Ntrain)]
sel_test  = myRange[Ntrain:N]
```

```python
# Dimension of the input and ARD
input_dim = X.shape[1]
```

```python
# kernels and model
Kbias   = GPy.kern.Bias(input_dim)
Kard    = GPy.kern.RBF(input_dim, variance=1, lengthscale=10,ARD=1)
Klinear = GPy.kern.Linear(input_dim,ARD=1)

k     = Kbias + Klinear

# model
model_translation = GPy.models.GPRegression(X[sel_train,],Y_translation_log[sel_train,],kernel = k)
#experimental_variation = var(errors[errors<threshold])
#model_translation.Gaussian_noise.variance.constrain_bounded(0.01,1)
model_translation
```

```python
# model optimization
model_translation.optimize(optimizer='bfgs', max_iters=200, messages=1)
```

```python
# Print the model 
model_translation
```

```python
## save the model
model_translation.pickle('model_translation')
```

```python
# Random generation of new genes according to the design rules
for j in range(0,Ngenes):
    gene_id = 'any'
    gene_seq = S[sel_train[j]]
    gene_transcription = Y1[sel_train[j]]
    gene_translation = Y2[sel_train[j]]
    gene_features = X[sel_train[j],]
    for k in range(Nseq):
        new_seq = generate_random_sequence(gene_seq) 
        Mevals[k][0]= eval(new_seq,gene_id,x_opt,scales)
        xx = extract_features(new_seq,gene_id).ix[:,1]
        x = np.array((xx-X0.mean(0))/X0.std(0))
        ymeta = np.vstack([ np.zeros((1,1)), np.ones((1,1))]) .astype(int)
        newX = np.hstack( [ np.vstack([x,x]), ymeta]).astype(float)
        Mevals[k][1:3] =  model.predict(newX,Y_metadata={'output_index':ymeta},full_cov=True)[0].T
        Mevals[k][3:5] =  model.predict(newX,Y_metadata={'output_index':ymeta},full_cov=True)[1].diagonal()
    Mgenes[j,0:5] = Mevals[np.argmin(Mevals[:,0])]
    Mgenes[j,5] = gene_transcription
    Mgenes[j,6] = gene_translation

means_pred_genes = (Mgenes[:,1]+Mgenes[:,2])/2
means_true_genes = (Mgenes[:,5]+Mgenes[:,6])/2
sdev_true_genes =  (np.zeros(10))/2
sdev_pred_genes = np.sqrt(Mgenes[:,3])/2+np.sqrt(Mgenes[:,4])/2
```
