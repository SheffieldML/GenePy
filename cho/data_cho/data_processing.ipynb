## Notebook to merge the raw data files of the project 
## We create a venn diagram to analise how the ids of the files overlapp

```python
# --- load requiered packages for the analysis
import numpy as np
import pandas as pd
```

```python
# --- mapping (note: the mapping id loaded nicely!)
df_mapping  = pd.read_csv('mapping_table.csv', sep=',').dropna()
df_mapping['cds'] = 'cds' + df_mapping['cds'].astype(str)
df_mapping['rna'] = 'rna' + df_mapping['rna'].astype(str)
```

```python
# Datasets (outputs of the model)
# --- RNA
df_rna = pd.read_csv('./data/Galaxy116.csv', sep='\t').dropna()
ids_rna = df_rna['tracking_id'].reset_index(drop=True)

# --- retain only rnas in rna_ids
for i in range(len(ids_rna)):
    if ids_rna[i][0]!='r':
        ids_rna = ids_rna.drop(i)
ids_rna = ids_rna.reset_index(drop=True)
ids_rna.to_csv('rna_id.csv',index=False)
```

```python
# --- Rates
df_rates = pd.read_csv('./data/p2_h2l_curated.csv',sep='\t',na_values='NA')
ids_rates = df_rates['ID']

# --- retain only the cds and translate to rnas
for i in range(len(ids_rates)):
    if ids_rates[i][0:3]!='cds':
        ids_rates = ids_rates.drop(i)
    else:
        df_aux = df_mapping['rna'][df_mapping['cds']==ids_rates[i]].reset_index(drop=True)
        if not df_aux.empty:
            ids_rates[i] = str(df_aux[0])
        else:
            ids_rates = ids_rates.drop(i)
ids_rates = ids_rates.reset_index(drop=True)
ids_rates.to_csv('rates_id.csv',index=False)
```

```python
# --- We extract the ids from the sequences
ids_5utr = pd.read_csv('./sequences/5utr_id.csv',sep='\t',na_values='NA',header=None,names=['ids'])
ids_3utr = pd.read_csv('./sequences/3utr_id.csv',sep='\t',na_values='NA',header=None,names=['ids'])

ids_cds  = pd.read_csv('./sequences/cds_id.csv',sep='\t',na_values='NA',header=None,names=['ids'])
ids_notranslated_cds  = pd.read_csv('./sequences/cds_notranslated_id.csv',sep='\t',na_values='NA',header=None)

ids_genes = pd.merge(pd.merge(ids_5utr, ids_3utr,on=['ids'], how='inner'), ids_cds, on=['ids'], how='inner')
```

```python
ids_genes.to_csv('gene_ids.csv',header=False, index=False)
```
