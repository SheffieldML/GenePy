# Analysis of promoter data

```python
# load requiered packages
%pylab inline
import numpy as np
import GPy
from matplotlib import pyplot as plt
import scipy.stats.stats 
from numpy import genfromtxt
from pandas import DataFrame, concat
import pandas as pd
from matplotlib.pyplot import *
from scipy.stats.stats import pearsonr as cor
from sklearn.preprocessing import scale
from pylab import pcolor, show, colorbar, xticks, yticks
```

```python
data = pd.read_csv('data.csv', sep='\t').dropna()
Y = data['Activity'].astype('Float64').as_matrix().reshape(145,1)
X = data.iloc[:,2:].astype('Float64').as_matrix()
```

```python
# Select sample for training and testin
N = X.shape[0]
input_dim = X.shape[1]

N_train = 100
N_test = N_train
myRange = range(N)
np.random.shuffle(myRange)
```

```python
# Index of the training and test genes
sel_train = myRange[0:(N_train)]
sel_test  = myRange[N_train:N]
```

```python
Y.shape
```

```python
# Model

# kernels and model
Kbias   = GPy.kern.Bias(input_dim)
Krbf = GPy.kern.RBF(input_dim,ARD=1)
k     = Kbias + Krbf

# model
model = GPy.models.GPRegression(X[sel_train,:],Y[sel_train,:],kernel=k)
model.optimize_restarts(20)
```

```python
model
```

```python
model.add.rbf.lengthscale
```

```python
# prediction

## prediction of the model in the training and test sets
prediction_train = model.predict(X[sel_train,:])
prediction_test  = model.predict(X[sel_test,:])

cor_train = cor(prediction_train[0],Y[sel_train,])[0]
cor_test = cor(prediction_test[0],Y[sel_test,])[0]

# Train data
plt.figure(figsize=(15,6))
plt.subplot(121)
plt.scatter(Y[sel_train,],prediction_train[0],c=prediction_train[1],s=50,alpha=25)
plt.xlabel('Observed',fontsize=20)
plt.ylabel('Predicted',fontsize=20)
plt.title('Train',fontsize=25)
plt.xlim(0, 100)
plt.ylim(0,  100)
plt.plot(np.linspace(0,100,100), np.linspace(0,100,100), 'b--')
plt.colorbar()
plt.text(30,6,'R = %2f' % cor_train,fontsize=20)


# Test data
plt.subplot(122)
plt.scatter(Y[sel_test,],prediction_test[0],c=prediction_test[1],s=50,alpha=25)
plt.title('Test',fontsize=25)
plt.xlabel('Observed',fontsize=20)
plt.ylabel('Predicted',fontsize=20)
plt.colorbar()
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.plot(np.linspace(0,100,100), np.linspace(0,100,100), 'b--')
plt.text(30,6,'R = %2f' % cor_test,fontsize=20)
```

```python
## Design of a batch of promoters
import GPyOpt

objective = GPyOpt.fmodels.experiments2d.branin() ## fake one
bounds = [(0,7),(0,6),(0,6),(0,6),(0,7),(0,5),(0,7),(0,5),(0,5),(0,4),(0,3),(0,4)]

BO_demo_parallel = GPyOpt.methods.BayesianOptimization(f= objective,
                                                       bounds=bounds,
                                                       X=X,
                                                       Y=-Y,
                                                       kernel=k)
```

```python
BO_demo_parallel.run_optimization(0,acqu_optimize_method='DIRECT',batch_method='lp',n_inbatch=10)
```

```python
# recompute the model
model = GPy.models.GPRegression(X,Y,kernel=k)
model.optimize_restarts(10)
```

```python
simulated_promoters = X
n_sim_promoters = 10000
uB = X.max(0)

for i in range(n_sim_promoters):
    x_new = (np.random.rand(input_dim)*uB).round()
    simulated_promoters = np.vstack((simulated_promoters,x_new ))
```

```python
## prediction with the GP model and index
Y_pred = model.predict(simulated_promoters)
Y_predflat = Y_pred[0].flatten()
index = Y_predflat.argsort()
data_index = [i for i,x in enumerate(index) if x <X.shape[0] ]
```

```python
plt.figure(figsize=(15,10))
plt.plot(Y_pred[0][index],label='Promoters expected activity')
plt.plot(Y_pred[0][index]+2*np.sqrt(Y_pred[1][index]),'b-.',label='Predicted error')
plt.plot(Y_pred[0][index]-2*np.sqrt(Y_pred[1][index]),'b-.')
plt.plot(data_index,Y_pred[0][index][data_index],'r.',markersize=10,label='Experiments' )
plt.ylabel('Activity',fontsize=20)
plt.xlabel('Promoter design index',fontsize=20)
plt.title('Promoter design curve',fontsize=25)
plt.ylim((-10,120))
plt.xlim((0,simulated_promoters.shape[0]))
plt.legend(loc=2)
```

```python
import pandas as pd
col_names = ['Predicted_activity','Variance','A', 'B', 'C', 'D','E', 'F', 'G','H','I','J','K','L']
predictions = np.hstack((Y_pred[0][index],np.sqrt(Y_pred[1][index])))

df_insilico_promoters = pd.DataFrame(np.hstack((predictions,simulated_promoters)),columns = col_names)
```

```python
df_insilico_promoters.to_csv('insilico_promoters.csv')
```
